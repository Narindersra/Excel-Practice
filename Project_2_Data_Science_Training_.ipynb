```python
Q1 = df['Damaged_Items'].quantile(0.25)
Q3 = df['Damaged_Items'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

high_damaged_items = df[df['Damaged_Items'] > upper_bound]

if not high_damaged_items.empty:
    print(f"Products with unusually high damaged item counts (above {upper_bound:.2f}):")
    display(high_damaged_items[['Product_Name', 'Damaged_Items', 'Inventory_Value']])
else:
    print("No products found with unusually high damaged item counts based on the IQR method.")
```

Identify products with unusually high damaged item counts and assess their impact on inventory management.

```python
# Filter products with unusually high damaged items using the previously calculated upper_bound
high_damaged_df = df[df['Damaged_Items'] > upper_bound].copy()

# Calculate the financial loss due to damaged items
high_damaged_df['Financial_Loss'] = high_damaged_df['Damaged_Items'] * high_damaged_df['Unit_Price']

# Calculate the percentage of stock that is damaged
high_damaged_df['Damage_Rate'] = (high_damaged_df['Damaged_Items'] / (high_damaged_df['Quantity_Stock'] + high_damaged_df['Damaged_Items'])) * 100

# Sort by financial loss to show impact
high_damaged_df = high_damaged_df.sort_values(by='Financial_Loss', ascending=False)

print(f"Assessment of {len(high_damaged_df)} products with high damage counts:")
display(high_damaged_df[['Product_Name', 'Category', 'Damaged_Items', 'Damage_Rate', 'Financial_Loss']])

total_loss = high_damaged_df['Financial_Loss'].sum()
print(f"\nTotal potential financial loss from these products: {total_loss:,.2f}")
```

Identify products that are approaching their expiry date and require immediate attention.

```python
import datetime

# Defining 'now' as the most recent purchase date in the dataset for relative context
reference_date = df['Purchase_Date'].max()
expiry_threshold = reference_date + datetime.timedelta(days=30)

# Filter products expiring soon
expiring_soon_df = df[(df['Expiry_Date'] >= reference_date) & (df['Expiry_Date'] <= expiry_threshold)].copy()

# Also identify already expired products
already_expired_df = df[df['Expiry_Date'] < reference_date].copy()

print(f"Reference Date (Latest Purchase): {reference_date.date()}")
print(f"Products expiring within 30 days (by {expiry_threshold.date()}):")

if not expiring_soon_df.empty:
    display(expiring_soon_df[['Product_Name', 'Category', 'Expiry_Date', 'Quantity_Stock', 'Status']].sort_values(by='Expiry_Date'))
else:
    print("No products are expiring within the next 30 days.")

if not already_expired_df.empty:
    print(f"\nALERT: {len(already_expired_df)} products have already expired.")
    display(already_expired_df[['Product_Name', 'Category', 'Expiry_Date', 'Quantity_Stock']].head(5))
```

Analyze the relationship between stock quantity and monthly sales to identify overstocked and understocked products.

```python
# Calculate Stock-to-Sales Ratio
# We add a small epsilon to avoid division by zero if sales were 0
df['Stock_to_Sales_Ratio'] = df['Quantity_Stock'] / (df['Sales_Last_Month'] + 1e-9)

# Define thresholds
# Overstocked: More than 3 months of stock relative to last month's sales
# Understocked: Less than 0.5 months of stock relative to last month's sales

overstocked_threshold = 3.0
understocked_threshold = 0.5

def categorize_stock(ratio):
    if ratio > overstocked_threshold:
        return 'Overstocked'
    elif ratio < understocked_threshold:
        return 'Understocked'
    else:
        return 'Balanced'

df['Stock_Efficiency'] = df['Stock_to_Sales_Ratio'].apply(categorize_stock)

# Summarize results
stock_summary = df['Stock_Efficiency'].value_counts().reset_index()
stock_summary.columns = ['Status', 'Count']

print("Stock Level Analysis Summary:")
display(stock_summary)

print("\nTop 5 Overstocked Products (High Capital Tie-up):")
display(df[df['Stock_Efficiency'] == 'Overstocked'].sort_values(by='Stock_to_Sales_Ratio', ascending=False)[['Product_Name', 'Category', 'Quantity_Stock', 'Sales_Last_Month', 'Stock_to_Sales_Ratio']].head(5))

print("\nTop 5 Understocked Products (High Stockout Risk):")
display(df[df['Stock_Efficiency'] == 'Understocked'].sort_values(by='Stock_to_Sales_Ratio')[['Product_Name', 'Category', 'Quantity_Stock', 'Sales_Last_Month', 'Stock_to_Sales_Ratio']].head(5))
```
